## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
#remove dulplicates
df = df.drop_duplicates()
#handling missing values
#separate numerical and categorical columns
num_cols = df.select_dtypes(include=['int64','float64']).columns
cat_cols = df.select_dtypes(include="object").columns
#fill numerical columns with median
for col in num_cols:
  df[col] = df[col].fillna(df[col].median())
#fill categorical columns with mode
for col in cat_cols:
  df[col] = df[col].fillna(df[col].mode()[0])


````

````python
df.nunique()
````

````python
df.isnull().sum()
````

### What all manipulations have you done and insights you found?



*  duplicate values were removed
*  missing values were filled
        1. numerical columns were filled using median to reduce the outliers
        2. categorical columns were filled using median to retain the most frequent category     
*   after cleaning , the dataset become more reliable and ready for analysis
