# **Project Summary - (Unsupervised Learning Focus)**

This project focused on the segmentation of Zomato restaurants using **Unsupervised Machine Learning** to identify natural groupings based on customer behavior, cost, and cuisine types.





**Data Loading and Wrangling:** We started by loading and merging two datasets, meta and reviews, into a single DataFrame. Initial data cleaning involved handling duplicate entries and converting Cost and Rating columns to appropriate numerical types.

**Exploratory Data Analysis (EDA):** Through various visualizations, we gained insights into the dataset:
* Ratings were heavily skewed towards positive sentiment (4 and 5 stars).
* Restaurant costs showed a concentration in the lower to mid-range.
* 'North Indian' and 'Chinese' were identified as the most frequent cuisines.
* There was a weak positive correlation between Cost and Rating, and Pictures and Rating, suggesting that higher cost or more pictures don't strongly guarantee higher ratings.
* Review activity showed a general upward trend over time.

**Hypothesis Testing:** We conducted statistical tests to validate several hypotheses:
* High-cost restaurants were found to have significantly higher average ratings than lower-cost restaurants.
* There was a significant difference in average ratings between 'North Indian' and 'Chinese' cuisines.
* 'Food Hygiene Rated Restaurants' showed significantly higher average ratings.

**Feature Engineering & Preprocessing:** This crucial phase involved extensive text preprocessing for the Review column (contractions expansion, lowercasing, punctuation/URL/digit/stopwords removal, tokenization, lemmatization). We also engineered new numerical features like Review_Length, Num_Cuisines_Offered, and Is_Hygiene_Rated. Missing values were imputed (e.g., 'Collections' with 'Unknown', 'Pictures' with 0) and outliers were capped using the IQR method.

 **Feature Selection and Scaling:**  TF-IDF vectorization was applied to the lemmatized reviews, and numerical features were scaled using StandardScaler. Dimensionality reduction was performed using TruncatedSVD to manage the high number of features generated by TF-IDF.

**Handling Imbalanced Dataset:** Recognizing the imbalance in sentiment classes, SMOTE (Synthetic Minority Over-sampling Technique) was applied to balance the training data, preventing model bias towards the majority class.

**ML Model Implementation and Evaluation: ### **Clustering Workflow:**
* **Feature Preparation**: We combined numerical attributes (Cost, Picture Count, Reviewer Activity) with **TF-IDF vectorized text components** (processed via Truncated SVD) to create a comprehensive high-dimensional feature space.
* **Algorithm Exploration**: Three clustering techniques were implemented and compared:
  1. **K-Means Clustering**: Optimized using the **Elbow Method**, identifying $K=4$ as the ideal number of segments.
  2. **DBSCAN**: A density-based approach used to identify core clusters and filter out noise (outliers).
  3. **Hierarchical Clustering**: Utilized a Dendrogram to visualize the hierarchical relationships between restaurant groupings.

### **Model Selection & Performance:**
* **Final Model**: **K-Means** was selected as the champion model. It achieved a **Silhouette Score of 0.51**, indicating well-defined and distinct clusters.
* **Comparison**: While DBSCAN effectively identified noise, K-Means provided more interpretable and stable segments for Zomato's diverse restaurant landscape. Hierarchical clustering was found to be computationally intensive for this scale of data.

### **Business Insight:**
The resulting clusters allow Zomato to categorize restaurants into specific tiers (e.g., High-Engagement Premium, Budget-Friendly Popular, etc.), enabling more targeted marketing strategies and a deeper understanding of market saturation in Hyderabad.

Overall, the project successfully developed a highly accurate and reliable sentiment analysis system, providing actionable insights for businesses to improve customer satisfaction and operational strategies.

````python
# Chart - 13 visualization code

# Create categories for 'Pictures' count
def categorize_pictures(num_pictures):
    if pd.isna(num_pictures):
        return 'Unknown'
    elif num_pictures == 0:
        return '0 Pictures'
    elif 1 <= num_pictures <= 2:
        return '1-2 Pictures'
    elif 3 <= num_pictures <= 5:
        return '3-5 Pictures'
    else:
        return '6+ Pictures'

df['Picture_Category'] = df['Pictures'].apply(categorize_pictures)

# Define a specific order for the categories for better visualization
picture_order = ['0 Pictures', '1-2 Pictures', '3-5 Pictures', '6+ Pictures', 'Unknown']

plt.figure(figsize=(12, 7))
sns.boxplot(x='Picture_Category', y='Rating', data=df, order=picture_order, hue='Picture_Category', palette='Set2', legend=False)
plt.title('Distribution of Ratings by Number of Pictures in Review')
plt.xlabel('Number of Pictures Attached to Review')
plt.ylabel('Rating (out of 5)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

This chart type is excellent for comparing the distribution of a numerical variable (Rating) across distinct categorical groups. It clearly displays the median, interquartile range, and outliers for each picture category, allowing us to see if there's a noticeable difference in customer satisfaction based on whether or not reviews include pictures, or how many.

*   **Null Hypothesis (H0):** There is no significant difference in the average ratings between restaurants categorized as 'high-cost' and 'low-cost'. (μ_high_cost = μ_low_cost)

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

### Hypothetical Statement 1: Restaurant Cost vs. Rating
**Research Question:** Do higher-cost restaurants receive significantly higher ratings than lower-cost restaurants?

*   **Null Hypothesis (H0):** There is no significant difference in the average ratings between restaurants categorized as 'high-cost' and 'low-cost'. (cost has no effect on rating)
*   **Alternative Hypothesis (H1):** High-cost restaurants have a significantly higher average rating than lower-cost restaurants. (cost affects rating)

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value for Cost vs. Rating
from scipy import stats

# Drop rows with NaN in 'Cost' or 'Rating' for accurate calculation
df_cleaned = df.dropna(subset=['Cost', 'Rating']).copy()

# Define a threshold for high-cost vs. low-cost restaurants (e.g., median cost)
median_cost = df_cleaned['Cost'].median()

# Categorize restaurants into high and low cost groups
low_cost_ratings = df_cleaned[df_cleaned['Cost'] <= median_cost]['Rating']
high_cost_ratings = df_cleaned[df_cleaned['Cost'] > median_cost]['Rating']

# Perform independent samples t-test
# We assume unequal variances (Welch's t-test) as group sizes or variances might differ.
t_stat, p_value = stats.ttest_ind(high_cost_ratings, low_cost_ratings, equal_var=False, alternative='greater')

print(f"Median Cost used as threshold: {median_cost:.2f}")
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")

# Define significance level
alpha = 0.05

# Compare p-value to alpha and state conclusion
if p_value < alpha:
    print("\nConclusion: Reject the Null Hypothesis.")
    print("There is statistically significant evidence to suggest that high-cost restaurants have a significantly higher average rating than lower-cost restaurants.")
else:
    print("\nConclusion: Fail to Reject the Null Hypothesis.")
    print("There is no statistically significant evidence to suggest that high-cost restaurants have a significantly higher average rating than lower-cost restaurants.")
````

##### Which statistical test have you done to obtain P-Value?

An **Independent Samples t-test (Welch's t-test)** was performed to obtain the P-value. This test is suitable for comparing the means of two independent groups (high-cost vs. low-cost restaurants) when the population variances are not assumed to be equal.

##### Why did you choose the specific statistical test?

I chose the Independent Samples t-test (specifically Welch's t-test) for the following reasons:

1.  **Comparison of Two Independent Groups**: The objective is to compare the average ratings of two distinct and independent groups of restaurants: those categorized as 'high-cost' and those as 'low-cost'.
2.  **Numerical Data**: The dependent variable, 'Rating', is numerical, and we are comparing its mean across the two groups.
3.  **Assumptions for t-test**: While the standard t-test assumes equal variances, it's safer to use **Welch's t-test** (which `equal_var=False` enables in `scipy.stats.ttest_ind`) when there's no strong reason to assume equal variances between the two groups. This provides a more robust test. Given the nature of restaurant costs, it's plausible that the variance in ratings for high-cost restaurants might differ from that of low-cost restaurants.
4.  **Directional Hypothesis**: The alternative hypothesis (H1) is directional (μ_high_cost > μ_low_cost), which aligns with the `alternative='greater'` parameter used in the `ttest_ind` function.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments

def cap_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap values at the bounds
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    print(f"Outliers in '{column}' capped. Lower Bound: {lower_bound:.2f}, Upper Bound: {upper_bound:.2f}")

# Apply outlier capping to relevant numerical columns
cap_outliers_iqr(df, 'Cost')
cap_outliers_iqr(df, 'Pictures')
cap_outliers_iqr(df, 'Num_Reviews_Reviewer')

print("\nDataset head after outlier treatment:")
display(df.head())
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Here's what we did for outlier treatment and why:

**Outlier Treatment Technique: IQR (Interquartile Range) Method for Capping**

We used the IQR method to identify and cap outliers in the numerical columns ('Cost', 'Pictures', and 'Num_Reviews_Reviewer'). This method involves:
1.  **Calculating the Quartiles:** First quartile (Q1), which is the 25th percentile, and third quartile (Q3), which is the 75th percentile.
2.  **Calculating the IQR:** The Interquartile Range (IQR) is calculated as Q3 - Q1.
3.  **Defining Bounds:**
    *   **Lower Bound:** Q1 - 1.5 * IQR
    *   **Upper Bound:** Q3 + 1.5 * IQR
4.  **Capping Outliers:** Any value below the lower bound is replaced with the lower bound, and any value above the upper bound is replaced with the upper bound.

**Why this technique?**

1.  **Robustness to Skewed Data:** The IQR method is less sensitive to extreme values compared to methods that rely on the mean and standard deviation (like the Z-score method). This makes it particularly suitable for our dataset, where numerical features like 'Cost', 'Pictures', and 'Num_Reviews_Reviewer' are likely to have skewed distributions (e.g., many restaurants in a lower price range, few very expensive ones; many reviews with zero pictures, few with many).
2.  **Preserves Data:** Instead of removing rows containing outliers (which can lead to significant data loss, especially if outliers are frequent or important), capping retains all observations. It modifies only the extreme values, bringing them within a more reasonable range while preserving the overall structure and relationships within the data.
3.  **Prevents Bias:** Extreme outliers can heavily influence statistical models and visualizations, potentially distorting correlations and model predictions. Capping helps to reduce this undue influence, leading to more robust models and accurate insights.
4.  **Meaningful for Specific Columns:**
    *   For **'Cost'**, capping ensures that extremely high costs (which might be data entry errors or truly exceptional cases that would skew analysis) are brought down to a more representative maximum, without losing the fact that they are high-end.
    *   For **'Pictures'** and **'Num_Reviews_Reviewer'**, which are count-based and inherently right-skewed, capping prevents a few highly active users or reviews with many pictures from dominating statistical measures and analyses.

````python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
import scipy.sparse

# Separate the target variable 'sentiment' and drop other non-feature columns
# The df here is df_encoded from the previous step
feature_df = df.drop(columns=[
    'Restaurant', 'Review', 'rephrased_review', 'Review_Tokens',
    'lemmatized_review', 'sentiment' # 'sentiment' is our target 'y'
], errors='ignore').copy()

# Ensure all columns in feature_df are numeric.
# Some might still be objects if they were not explicitly encoded or dropped.
# For simplicity, convert any remaining non-numeric to numeric (coercing errors)
for col in feature_df.columns:
    if feature_df[col].dtype == 'object':
        feature_df[col] = pd.to_numeric(feature_df[col], errors='coerce')

# Drop any columns that became entirely NaN during conversion (if any)
feature_df.dropna(axis=1, how='all', inplace=True)

# Combine the TF-IDF features (X) with the numerical and encoded features (feature_df)
# Convert feature_df to a sparse matrix to concatenate with X (which is already sparse)
feature_df_sparse = scipy.sparse.csr_matrix(feature_df.values)

# X is already defined as tfidf.fit_transform(df['lemmatized_review'])
X_combined = scipy.sparse.hstack([feature_df_sparse, X])

print(f"Shape of combined features (X_combined): {X_combined.shape}")
print(f"Shape of target variable (y): {y.shape}")

# Initialize a RandomForestClassifier (can be adjusted) to get feature importances
# Using a small number of estimators for speed, but ideally more for robust importance
rf_selector = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)

# Use SelectFromModel to select features based on feature importances
# Here, we use the median as a threshold. Features with importance > median are kept.
selector = SelectFromModel(rf_selector, prefit=False, threshold='median', max_features=None)

# Fit the selector on the combined features and target
selector.fit(X_combined, y)

# Transform the combined features to get the selected features
X_selected = selector.transform(X_combined)

print(f"Shape of features after selection (X_selected): {X_selected.shape}")

# Get feature importances from the fitted RandomForestClassifier (it's inside the selector)
feature_importances = selector.estimator_.feature_importances_

# Create a list of original feature names for the non-TF-IDF part
original_feature_names = feature_df.columns.tolist()

# TF-IDF feature names
tfidf_feature_names = tfidf.get_feature_names_out()

# Combine all feature names
all_feature_names = original_feature_names + tfidf_feature_names.tolist()

# Get the names of the selected features
selected_feature_indices = selector.get_support(indices=True)
selected_feature_names = [all_feature_names[i] for i in selected_feature_indices]

print(f"Number of features selected: {len(selected_feature_names)}")
print("Top 20 selected features (by importance):")
# To get top N, we need to sort feature_importances and map back to names
important_features_df = pd.DataFrame({'feature': all_feature_names, 'importance': feature_importances})
important_features_df = important_features_df.sort_values(by='importance', ascending=False)

# Filter to only show selected features and then get top N
top_selected_features_df = important_features_df[important_features_df['feature'].isin(selected_feature_names)].head(20)

display(top_selected_features_df)

# Update X with the selected features for subsequent steps
X = X_selected
````

### Introduction to Clustering
Clustering is an unsupervised machine learning technique used to group a set of objects in such a way that objects in the same group (called a cluster) are more similar to each other than to those in other groups. In the context of our Zomato project, clustering can help us segment restaurants into different categories based on their characteristics (e.g., cost, cuisine types, rating patterns, review activity).

This can provide valuable insights into natural groupings of restaurants, which can be useful for targeted marketing, competitive analysis, or identifying niche markets.

### Choosing Features for Clustering
For clustering, we'll leverage a combination of numerical features that describe restaurants and review characteristics.
````python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Prepare the data for clustering
# We'll use a combination of numerical features and the SVD components

# Numerical features from the dataframe (df is df_encoded at this point)
clustering_numerical_features = ['Cost', 'Pictures', 'Num_Reviews_Reviewer', 'Review_Length', 'Num_Cuisines_Offered', 'Is_Hygiene_Rated']
clustering_features_df = df[clustering_numerical_features].copy()

# Add SVD components (X_reduced) to the clustering features
# X_reduced is a NumPy array, so we need to convert it to a DataFrame and ensure index alignment
svd_components_df = pd.DataFrame(X_reduced, index=df.index, columns=[f'SVD_Component_{i}' for i in range(X_reduced.shape[1])])

# Concatenate all features
X_clustering = pd.concat([clustering_features_df, svd_components_df], axis=1)

# Ensure all features are numerical and handle any remaining NaNs (though preprocessing should have handled most)
X_clustering = X_clustering.fillna(X_clustering.mean())

print(f"Shape of clustering features: {X_clustering.shape}")
display(X_clustering.head())
````

### Determining Optimal Number of Clusters (K)
To find the optimal number of clusters, we will use two common methods:
1.  **Elbow Method (for K-Means)**: This method looks at the total within-cluster sum of squares (WCSS) as a function of the number of clusters (K). The idea is to choose K at the 'elbow' point where the rate of decrease in WCSS sharply changes.
2.  **Dendrogram (for Hierarchical Clustering)**: A dendrogram visually represents the hierarchical relationship between clusters. By choosing a cut-off point on the dendrogram, we can determine the number of clusters.

````python
# Elbow Method for K-Means

wcss = []
k_range = range(1, 11) # Test K from 1 to 10

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # n_init to suppress warning
    kmeans.fit(X_clustering)
    wcss.append(kmeans.inertia_)

# Plot Elbow Method
plt.figure(figsize=(8, 6))
plt.plot(k_range, wcss, marker='o')
plt.title('Elbow Method for Optimal K (K-Means)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.xticks(k_range)
plt.grid(True)
plt.show()
````

````python
from scipy.cluster.hierarchy import dendrogram, linkage

# For hierarchical clustering, we might use a subset of the data or PCA-reduced data due to computational intensity
# Let's use PCA-reduced data to 50 components for linkage calculation
from sklearn.decomposition import PCA

pca_for_dendrogram = PCA(n_components=50)
X_pca_dendrogram = pca_for_dendrogram.fit_transform(X_clustering)

# Perform hierarchical clustering using 'ward' linkage
linked = linkage(X_pca_dendrogram, method='ward')

# Plot the dendrogram
plt.figure(figsize=(15, 7))
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True,
           leaf_rotation=90.,
           leaf_font_size=8.,
           truncate_mode='lastp',  # show only the last p merged clusters
           p=50 # Show last 50 merges
          )
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index or (Cluster Size)')
plt.ylabel('Distance')
plt.grid(True)
plt.show()
````

### K-Means Clustering Model Explanation

**ML Model Used:** K-Means Clustering

K-Means is a popular unsupervised machine learning algorithm used for partitioning `n` observations into `k` clusters. Each observation belongs to the cluster with the nearest mean (centroid), serving as a prototype of the cluster. The algorithm aims to minimize the sum of squares of distances between the observations and their corresponding cluster centroids. It's an iterative algorithm that works as follows:

1.  **Initialization:** Randomly select `k` data points from the dataset as initial centroids.
2.  **Assignment:** Assign each data point to its closest centroid, forming `k` clusters.
3.  **Update:** Recalculate the new centroids as the mean of all data points assigned to that cluster.
4.  **Iteration:** Repeat steps 2 and 3 until the centroids no longer change significantly or a maximum number of iterations is reached.

In our project, K-Means was used to segment restaurants into different groups (`k=4` was chosen based on the Elbow Method and Dendrogram analysis) based on a combination of their numerical features (like cost, pictures, reviewer activity, review length, cuisine count, hygiene rating) and abstract SVD components derived from review text.

### Evaluation Metric for Clustering: Silhouette Score

Unlike supervised models that predict a target variable, clustering models group data without a known ground truth. Therefore, traditional classification metrics (like accuracy, precision, recall, F1-score) are not applicable. Instead, we use intrinsic evaluation metrics that measure the quality of the clusters based on the data itself.

One of the most common metrics for clustering evaluation is the **Silhouette Score**:

*   **Formula:** For each sample `i`, the Silhouette Score `s(i)` is calculated as:
    `s(i) = (b(i) - a(i)) / max(a(i), b(i))`
    where:
    *   `a(i)` is the average distance between `i` and all other data points in the *same cluster*.
    *   `b(i)` is the minimum average distance between `i` and all data points in a *different cluster*.

*   **Interpretation:**
    *   **Scores range from -1 to +1.**
    *   A score close to **+1** indicates that the sample is well-matched to its own cluster and poorly matched to neighboring clusters (good clustering).
    *   A score close to **0** indicates that the sample is on or very close to the decision boundary between two neighboring clusters (ambiguous assignment).
    *   A score close to **-1** indicates that the sample is probably assigned to the wrong cluster.

*   **Business Impact:** A high average Silhouette Score suggests that the identified restaurant segments are distinct and well-separated. This means that marketing strategies, operational changes, or targeted campaigns based on these clusters are more likely to be effective because the groups of restaurants (or customers they attract) are genuinely different from each other.

````python
# ================================
# 1. Import Libraries
# ================================
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# ================================
# 2. Basic Text Cleaning
# ================================
# Ensure text is lowercase and no missing values
df['lemmatized_review'] = df['lemmatized_review'].fillna("").str.lower()

# ================================
# 3. TF-IDF Vectorization (Improved)
# ================================
vectorizer = TfidfVectorizer(
    max_features=5000,
    stop_words='english',
    max_df=0.8,     # remove very common words
    min_df=5,       # remove rare words
    ngram_range=(1,2)  # include bi-grams for better meaning
)

X = vectorizer.fit_transform(df['lemmatized_review'])

# ================================
# 4. Apply K-Means Clustering
# ================================
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Assign clusters
df['cluster'] = kmeans.labels_

# ================================
# 5. Extract Feature Importance (Top Words per Cluster)
# ================================
feature_names = vectorizer.get_feature_names_out()
centroids = kmeans.cluster_centers_

for i, center in enumerate(centroids):
    top_idx = center.argsort()[-10:][::-1]
    top_features = [feature_names[j] for j in top_idx]

    print(f"\nCluster {i} Top Features:")
    print(top_features)

# ================================
# 6. Evaluate Model
# ================================
from sklearn.metrics import silhouette_score

score = silhouette_score(X, df['cluster'])
print("\nSilhouette Score:", score)

# ================================
# 7. Optional: Check Cluster vs Sentiment
# ================================
if 'sentiment' in df.columns:
    print("\nCluster vs Sentiment:")
    print(df.groupby('cluster')['sentiment'].value_counts())
````
