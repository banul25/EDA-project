MODEL TRAINING

i used a pretrained YOLOV8s.pt(yolov8 small model) and trained it on my dataset, and results are stored in yolo_training_results

```python
!yolo task=detect mode=train model=yolov8s.pt data=/content/drive/MyDrive/object_detection_Dataset/data.yaml epochs=10 imgsz=320 batch=8 project=/content/drive/MyDrive/yolo_training_results
```

VALIDATION

```python
!yolo task=detect mode=val model=/content/drive/MyDrive/yolo_training_results/train2/weights/best.pt data=/content/drive/MyDrive/object_detection_Dataset/data.yaml
```

after training, i validated the model and got the following results
1)bird
   * precision 0.721
   * recall: 0.565
   * mAP(MEAN AVARAGE PRECICION):0.612
2)drone
  * precision: 0.879
  * recall: 0.876
  * mAP:0.903
  model performed well on drone than bird
